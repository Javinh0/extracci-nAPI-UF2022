# Desafío descargando datos de una API publica

A continuación se procederá a descargar la data de una API publica que entrega indicadores economicos, esto con el fin de obtener datos relacionados los parametros y evaluaciones de paises en el marco del desarrollo de las ODS con el fin de contribuir al informe final.

Url de la API: https://mindicador.cl/

### Resumen de la informacióna  obtener

Este es un servicio open source (web service) que entrega los principales indicadores económicos para Chile en formato JSON. Tanto los indicadores diarios como históricos para que desarrolladores puedan utilizarlos en sus aplicaciones o sitios web.

La aplicación mapea constantemente el sitio del Banco Central de Chile manteniendo así nuestra base de datos actualizada con los últimos valores del día.

### Obtención de valores UF durante el año 2022

```python
#Comenzamos con la importacion de librerias de Python y la solicitud de la información a la API según indica la documentación de esta. 
import requests
import json
import pandas as pd


url = 'https://mindicador.cl/api/uf/2022'

response = requests.get(url)

#consultamos previamente el la cantidad de items que se importaran
print(len(response.text))
response.text
```

```python
#se da formato al json
texto=response.text
json.loads(texto)
```

```python
#se crea la tabla de de nuestro df
jsondata=json.loads(texto)
df=pd.DataFrame.from_dict(jsondata)
df
```

### Proceso de limpieza y orden de datos

```python
# Cargar el JSON de nuevo
json_data = json.loads(texto)

# Crear un DataFrame nuevo para poder separar la columna de serie 
df2 = pd.json_normalize(json_data['serie'])

# Convertir la columna 'fecha' al formato de fecha eliminando las horas.
df2['fecha'] = pd.to_datetime(df2['fecha']).dt.date

# Renombrar las columnas según tus requerimientos
df2.rename(columns={'fecha': 'Fecha', 'valor': 'Valor'}, inplace=True)

# Mostrar el DataFrame resultante
df2
```

```python
# Eliminar la columna 'serie' de df
df.drop('serie', axis=1, inplace=True)

# Concatenar df y df2 a lo largo de las columnas (axis=1)
df = pd.concat([df, df2], axis=1)

df
```
